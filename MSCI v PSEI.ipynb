```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style('whitegrid')
plt.figure(figsize=(14, 16))

fig1 = sns_plot = sns.barplot(x=df2['Weight'], y=df2.index, hue=df2['Index'], palette='RdBu')
#fig1 = sns_plot.figure.suptitle('MSCI v PSEI', size=40)
#fig1 = plt.figure(1)
fig1.text(8,18, "PSEI", ha="right", va="top", size="40",color="#e49d8b")
fig1.text(9,18, "vs", ha="right", va="top", size="40")
fig1.text(11,18,"MSCI", ha="right", va="top", size="40",color="#9bc6d4")
plt.legend([])
fig1.yaxis.label.set_size(20)
fig1.xaxis.label.set_size(20)
fig1.tick_params(labelsize=25)
plt.xlabel('Weight%',fontsize = 25)
plt.xticks(fontsize = 20)  #,rotation=40
plt.yticks(fontsize = 14)
plt.savefig('MSCIPSEI.png')
plt.show()

```
